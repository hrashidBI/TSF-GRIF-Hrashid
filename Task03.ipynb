```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings(action='ignore')
from pylab import rcParams
```

```python
dset=pd.read_csv(r"/content/SampleSuperstore.csv")
dset.head()
```

```python
ndata=dset.drop(['Region'],axis=1)
```

```python
ndata.head()
```

```python
ndata.duplicated().sum()
```

```python
ndata.drop_duplicates(keep="first",inplace=True)
ndata.shape
```

```python
ndata.describe()
```

```python
correlation=ndata.corr()
```

```python
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True);
```

```python
sns.pairplot(ndata);
```

```python
plt.figure(figsize=(18,8))
plt.bar('Sub-Category','Category',data=ndata,color='b');
```

```python
ndata.groupby('Sub-Category')['Profit','Sales'].agg(sum).plot.bar()
plt.title(' Profit VS Sales per Sub-Category');
```

```python
ndata['Cost']=ndata['Sales']-ndata['Profit']
ndata['Cost'].head()
```

```python
ndata['Profit %']=(ndata['Profit']/ndata['Cost'])*100
```

```python
ndata.sort_values(['Profit %','Sub-Category'],ascending=False).groupby('Profit %').head(5)
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
sns.distplot(ndata['Sales'], color = 'g')
plt.title('The Distribution of Sales', fontsize = 30)
plt.xlabel('Range of Sales', fontsize = 15)
plt.ylabel('No. of Sales count', fontsize = 15)
plt.show()
```
